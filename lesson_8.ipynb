```python
def assign_grade(Rating):
    if Rating == 5:
        return 'Excellent'
    elif 3 <= Rating < 5:
        return 'Good'
    else:
        return 'Bad'
data['Grade'] = data['Rating'].apply(assign_grade)
data[['Customer', 'Rating', 'Grade']]

```
