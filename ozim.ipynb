```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('sales_dataa.xlsx')
df
```

```python
sns.countplot(x= df['Quantity'],hue=df['Category'],palette='muted')
```

```python
df['total']= df["Quantity"]*df['Price']
df
```

```python
df['Date']= pd.to_datetime(df['Date'])
df = df.sort_values(by=['Date'])
df
```

```python
sns.barplot(x=df['Category'],y= df['total'])
plt.xlabel('Mahsulotlar turi')
plt.ylabel('mahsulotlarga ishlatilgan mablag')
plt.show()
```

```python
sns.lineplot(x=df['Date'],y=df['total'],hue=df['Category'])
plt.xticks(rotation=75)
plt.grid(False)
plt.show()
```

```python
sns.boxplot(x=df["Product"],y=df['Price'],hue=df['Category'])
plt.xticks(rotation=75)
plt.show()
```

```python
pivot_df= pd.pivot_table(df,values='total',index='Date',columns='Category',aggfunc='sum')
plt.figure(figsize=(9,6))
sns.heatmap(pivot_df,annot=True,cmap='YlGnBu',linewidths=.5)

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('sales_dataa.xlsx')

sns.scatterplot(x=df['Product'], y=df['Price'], palette='cool', hue=df['Category'])
plt.xticks(rotation=75)
plt.grid(False)
plt.show()
```

```python
import pandas as pd
top = df.nlargest(10,"Price")
top

plt.figure(figsize=(9,6))
sns.barplot(x=top['Product'], y= top['Price'])
plt.title('Top 10 expensive products ')
plt.show()
```
